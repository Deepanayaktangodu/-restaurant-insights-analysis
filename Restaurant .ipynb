```python
# Center the map at a global average
map_center = [df["Latitude"].mean(), df["Longitude"].mean()]

# Create base map
restaurant_map = folium.Map(location=map_center, zoom_start=2)

# Add clusters
marker_cluster = MarkerCluster().add_to(restaurant_map)

# Add restaurant markers to the map
for idx, row in df.iterrows():
    popup_info = f"Cuisine: {row['Cuisines']}<br>Rating: {row['Aggregate rating']}"
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=popup_info,
        icon=folium.Icon(color="green", icon="cutlery", prefix='fa')
    ).add_to(marker_cluster)

# Save to HTML file
restaurant_map.save("restaurant_map.html")
print("Interactive map saved as restaurant_map.html")
```

2) Level 2 Task 4: Restaurant Chains

```python
# Count how many times each restaurant name appears
chain_counts = df["Restaurant Name"].value_counts()
restaurant_chains = chain_counts[chain_counts > 1]  # Chains with more than 1 outlet

# Create a DataFrame with average rating and vote count
chain_data = df[df["Restaurant Name"].isin(restaurant_chains.index)].groupby("Restaurant Name").agg({
    "Restaurant Name": "count",
    "Aggregate rating": "mean",
    "Votes": "mean"
}).rename(columns={"Restaurant Name": "Branch Count", "Aggregate rating": "Avg Rating", "Votes": "Avg Votes"}).round(2)

# Top chains by number of branches
top_chains = chain_data.sort_values("Branch Count", ascending=False).head(10)
print(top_chains)

# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(data=top_chains.reset_index(), x="Branch Count", y="Restaurant Name", palette="Set2")
plt.title("Top 10 Restaurant Chains by Branch Count")
plt.xlabel("Number of Branches")
plt.ylabel("Restaurant Chain")
plt.tight_layout()
plt.show()
```

3) Level 3 Task 1: Restaurant Review

```python
# 1. Frequency of Rating Text
rating_text_counts = df["Rating text"].value_counts()
print("Rating Text Distribution:\n", rating_text_counts)

# 2. Map Rating Text to Sentiment Categories
sentiment_map = {
    "Excellent": "Positive",
    "Very Good": "Positive",
    "Good": "Neutral",
    "Average": "Neutral",
    "Poor": "Negative",
    "Not rated": "Unknown"
}
df["Sentiment"] = df["Rating text"].map(sentiment_map)

# 3. Visualize Sentiment Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Sentiment", order=["Positive", "Neutral", "Negative", "Unknown"], palette="Set2")
plt.title("Distribution of Sentiment Based on Rating Text")
plt.xlabel("Sentiment")
plt.ylabel("Number of Restaurants")
plt.tight_layout()
plt.show()

# 4. Average Votes by Sentiment
avg_votes_by_sentiment = df.groupby("Sentiment")["Votes"].mean().round(2)
print("\nAverage Votes by Sentiment:\n", avg_votes_by_sentiment)

# 5. Simulate 'Review Length' using number of words in Rating Text (for demo only)
df["Review Length"] = df["Rating text"].apply(lambda x: len(x.split()))
review_vs_rating = df.groupby("Aggregate rating")["Review Length"].mean().round(2)

# Plot: Avg "Review Length" vs Rating
plt.figure(figsize=(8, 5))
sns.lineplot(x=review_vs_rating.index, y=review_vs_rating.values, marker="o")
plt.title("Simulated Review Length vs Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Simulated Review Length")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Since actual review text was not available, we adapted the task using the Rating text column to simulate customer sentiment.

Mapped each Rating text to a sentiment category: Positive, Neutral, Negative, or Unknown.

Calculated frequency distribution and average votes across these sentiment groups.

Simulated “review length” by measuring word count of Rating text (for correlation demo with rating).
